### 1) Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

```python
import numpy as np
import matplotlib.pyplot as plt

# input user
A = float(input("Enter coeff A (for x**4): "))
B = float(input("Enter coeff B (for x**3): "))
C = float(input("Enter coeff C (for x**2): "))
D = float(input("Enter coeff D (for x): "))
E = float(input("Enter coeff E (constant): "))

x = np.arange(-100, 101, 1)

y = A * (x**4) + B * (x**3) + C * (x**2) + D * x + E

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label=f'{A}x^4 + {B}x^3 + {C}x^2 + {D}x + {E}')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('4th Degree Polynomial or biquadratic Equation')
plt.legend()

# Show the plot
plt.show()
```

### 2) Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

### sales_data = {'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]}

### Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.

```python
import pandas as pd
import matplotlib.pyplot as plt

# make sales data dictionary
sales_data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],
    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]
}

# DataFrame conversion friom dictionary
df = pd.DataFrame(sales_data)

# compute total sales per product
total_sales = df.groupby('Product')['Sales'].sum()

# Create bar plot
plt.figure(figsize=(8, 6))
total_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=0)

for a, b in enumerate(total_sales):
    plt.text(a, b + 5, str(b), ha='center')

# Show the plot or graph
plt.tight_layout()
plt.show()

# Print the total sales
print("\nTotal sales by product:")
print(total_sales)
```

### 3)  Create visualizations for the following mathematical functions using Matplotlib: Plot the following single-variable functions over the range [10,10], and include a title and labels for the axes:

### (1) y = cos(x)

### (2) y = e^x

### (3) y = log(x), where x>0

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x ranges 
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
x3 = np.linspace(0.1, 10, 400)  # as we know  x > 0 for log(x), else we get undefined output 

# solve for   function values
y1 = np.cos(x1)
y2 = np.exp(x2)
y3 = np.log(x3)

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(6, 12))

# Plot y = cos(x)
axes[0].plot(x1, y1, label=r'$y = \cos(x)$', color='b')
axes[0].set_title("Plot of y = cos(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].legend()
axes[0].grid()

# Plot y = e^x
axes[1].plot(x2, y2, label=r'$y = e^x$', color='g')
axes[1].set_title("Plot of y = e^x")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].legend()
axes[1].grid()

# Plot y = log(x)
axes[2].plot(x3, y3, label=r'$y = \log(x)$', color='r')
axes[2].set_title("Plot of y = log(x)")
axes[2].set_xlabel("x")
axes[2].set_ylabel("y")
axes[2].legend()
axes[2].grid()

plt.tight_layout()
plt.show()
```

### Generate surface plots for these multi-variable functions over the range x=[10,10] and y=[10,10] , ensuring to add a title and labels for all axes:

### (1) z = cos(sqrt(x^2+y^2)

### (2) z = e^(-(x^2+y^2))

### (3) z =  log(x^2+y^2) where x^2+y^2>0

```python
import numpy as np
import matplotlib.pyplot as plt

# Set range for x and y
x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(x, y)

# Compute function values
Z_cos = np.cos(np.sqrt(X**2 + Y**2))  # Function 1: Cosine radial distance
Z_exp = np.exp(-(X**2 + Y**2))        # Function 2: Gaussian function

# Handling log(0) undefined case
X_safe, Y_safe = np.where(X**2 + Y**2 == 0, 1e-10, X), np.where(X**2 + Y**2 == 0, 1e-10, Y)
Z_log = np.log(X_safe**2 + Y_safe**2)  # Function 3: Log function

# Create figure for z = cos(sqrt(x**2 + y**2))
fig = plt.figure()
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z_cos, cmap='viridis')
ax1.set_title(r'$z = \cos(\sqrt{x^2 + y^2})$')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')

# Create figure for z = e**(-(x**2 + y**2))
fig = plt.figure()
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Z_exp, cmap='plasma')
ax2.set_title(r'$z = e^{-(x^2 + y^2)}$')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_zlabel('Z-axis')

# Create figure for z = log(x**2 + y**2)
fig = plt.figure()
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, Z_log, cmap='cividis')
ax3.set_title(r'$z = \log(x^2 + y^2)$')
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')
ax3.set_zlabel('Z-axis')
# print plot or graphs
plt.show()
```

### 4) Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of Optimization for Engineering Design: Algorithms and Examples by KALYANMOY DEB, 2nd edition
### f(x) = 3x**2 + 2x



```python
import numpy as np

def f(x):
    return 5*x**2 + 7*x

def central_difference(x, h=1e-5):
    first_derivative = (f(x + h) - f(x - h)) / (2 * h)
    second_derivative = (f(x + h) - 2*f(x) + f(x - h)) / (h**2)
    return first_derivative, second_derivative


x_test = 1
f_prime, f_double_prime = central_difference(x_test)
print(f"First derivative at x={x_test}: {f_prime}")
print(f"Second derivative at x={x_test}: {f_double_prime}")
```

### 4)  For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).



```python
import numpy as np

def J(w):
    return w**2 + (54 / w)  # Function to minimising

def find_min(a, b, n):
    for _ in range(n):
        mid = np.mean([a, b])  
        if J(a) > J(mid) < J(b):
            b = mid 
        else:
            a = mid
    return np.mean([a, b]) 
    
# create initial range and number of iterations
a, b, n = 1, 10, 20
min_w = find_min(a, b, n)

print("Approximate minimum w:", min_w)
```
